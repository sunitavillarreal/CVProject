```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
options = ChromeOptions()
driver = webdriver.Chrome(options=options)
```

```python
driver.get('https://images.google.com')
search_box = driver.find_element(By.NAME, "q")
#search_button = driver.find_element(By.NAME, "btnK")
```

```python
import time
import re
url_re = re.compile('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
#img_re = re.compile('(?<=/)[^/]+\.(?:jpe?g|png|gif|tiff?)') # 原图
img_re = re.compile('encrypted-tbn0\.gstatic\.com') # thumbnail

search_box.send_keys("immigration border" + Keys.ENTER)
action = webdriver.ActionChains(driver)
src = driver.page_source
prev_src = ''
wait = 1
wait_pgdn = 0.1
page_counter = 100

img_url_list = []
more_button = driver.find_element(By.CLASS_NAME, 'mye4qd')
img_n = -1

while page_counter > 0 and img_n < len(img_url_list):
    #print(page_counter)
    print(len(img_url_list))
    img_n = len(img_url_list)
    url_list = url_re.findall(src)
    for url in url_list:
        img_match = img_re.findall(url)
        if img_match:
            img_url_list.append(url)

    #more_button = driver.find_element(By.XPATH, '//button[text()="显示更多搜索结果"]')
    
    driver.execute_script("arguments[0].click();", more_button)
    pgdn_counter = 100
    #if more_button:
    #more_button.click()
    while pgdn_counter > 0:
        action.key_down(Keys.PAGE_DOWN).key_up(Keys.PAGE_DOWN).perform()
        pgdn_counter -= 1
        time.sleep(wait_pgdn)
    #while 
    
    time.sleep(wait)
    page_counter -= 1
    prev_src = src
    src = driver.page_source
    #else:
    #    break
#driver.quit()
    img_url_list = list(dict.fromkeys(img_url_list))
print(len(img_url_list))
```

```python
import os
import pathlib

current_dir = os.getcwd()
img_download_dir_name = 'img_download'
pathlib.Path(os.path.join(current_dir, img_download_dir_name)).mkdir(parents=True, exist_ok=True) 
img_download_dir = os.path.join(current_dir, img_download_dir_name)
```

```python
#import urllib.request
import urllib
import socket
socket.setdefaulttimeout(10)

img_filename_re = re.compile('(?<=/)[^/]+\.(?:jpe?g|png|gif|tiff?)')

total = len(img_url_list)
counter = 0
for i, img_url in enumerate(img_url_list):
    print(i + 1, '/', total)
    #img_file_name_match = img_filename_re.findall(img_url)
    img_file_name_match = True
    #print(str(os.path.join(img_download_dir, img_file_name)))
    #print(img_url)
    if img_file_name_match:
        #img_file_name = img_file_name_match[0]
        img_file_name = str(counter) + '.jpg'
        try:
            #urllib.request.urlretrieve(img_url, str(os.path.join(img_download_dir, img_file_name)))
            urllib.request.urlretrieve(img_url, str(os.path.join(img_download_dir, img_file_name)))
            print('Downloaded', img_url)
            counter += 1
        except urllib.error.HTTPError:
            print('HTTPError', img_url)
        except socket.timeout:
            print('Timeout', img_url)
print('Total downloaded:', counter)
```
